## The primary purpose of this project is to investigate and test a method for determining when a container has been emptied.
#### Model Training: With the preprocessed data and extracted features, we can train a machine learning model to predict determining when a container has been emptied. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('Smart_Bin.csv')
df.head()
```

```python
df['Class'].unique()
```

# Information overview of the dataset
### here is a brief explanation of each column:

- FL_B: This column could represent the amount of waste generated by a specific user or area.
- FL_A: This column could represent the amount of waste generated by another specific user or area.
- VS: This column could represent the volume of waste generated by a specific user or area.
- FL_B_3: This column could represent the amount of waste generated by a specific user or area in the last 3 days.
- FL_A_3: This column could represent the amount of waste generated by another specific user or area in the last 3 days.
- FL_B_12: This column could represent the amount of waste generated by a specific user or area in the last 12 days.
- FL_A_12: This column could represent the amount of waste generated by another specific user or area in the last 12 days.
- Container Type: This column could represent the type of waste container used in a specific area.
- Recyclable fraction: This column could represent the proportion of waste in a specific container that is recyclable.

#### Replace Missing Values

```python
def remove_column(column_name):
    df[column_name]  = df[column_name].fillna(df[column_name].median())
```

```python
i = 1
for i in range(1,8):
    remove_column(df.columns[i])
    i += 1
df
```

```python
df.isnull().sum()
```

```python
df.describe()
```

#### Normalizing target variable

```python
encoder = LabelEncoder()
df['Class'] = encoder.fit_transform(df['Class'])
df['Class'].unique()
```

```python
df['Container Type'].unique()
```

```python
df['Container_Type'] = encoder.fit_transform(df['Container Type'])
df['Container_Type'].unique()
```

```python
df['Recyclable fraction'].unique()
```

```python
df['Recyclable_fraction'] = encoder.fit_transform(df['Recyclable fraction'])
df['Recyclable_fraction'].unique()
```

#### Visualizations to check outliers

```python
import seaborn as sns

plt.figure(figsize=(20,15))
plt.subplot(2,4,1)
plt.title('class')
sns.boxplot(df['Class'])
plt.subplot(2,4,2)
plt.title('FL_B')
sns.boxplot(df['FL_B'])
plt.subplot(2,4,3)
plt.title('FL_A')
sns.boxplot(df['FL_A'])
plt.subplot(2,4,4)
plt.title('VS')
sns.boxplot(df['VS'])
plt.subplot(2,4,5)
plt.title('FL_B_3')
sns.boxplot(df['FL_B_3'])
plt.subplot(2,4,6)
plt.title('FL_A_3')
sns.boxplot(df['FL_A_3'])
plt.subplot(2,4,7)
plt.title('FL_B_12')
sns.boxplot(df['FL_B_12'])
plt.subplot(2,4,8)
plt.title('FL_A_12')
sns.boxplot(df['FL_A_12'])
plt.show()
```

#### Removing Outliers

```python
df['FL_B'].values[df['FL_B'].values > 100] = df['FL_B'].mean()
df['FL_B_3'].values[df['FL_B_3'].values > 100] = df['FL_B'].mean()
df['FL_B_12'].values[df['FL_B_12'].values > 100] = df['FL_B_12'].mean()
df['FL_A'].values[df['FL_A'].values > 100] = df['FL_A'].mean()
df['FL_A_3'].values[df['FL_A_3'].values > 100] = df['FL_A_3'].mean()
df['FL_A_12'].values[df['FL_A_12'].values > 100] = df['FL_A_3'].mean()
```

### Calculation of change in fill level

```python
df['FL_C'] = df['FL_A'] - df['FL_B']
df['FL_C_3'] = df['FL_A_3'] - df['FL_B_3']
df['FL_C_12'] = df['FL_A_12'] - df['FL_B_12']
```

```python
from matplotlib import rcParams
from matplotlib.cm import rainbow
```

```python
dfs = df.select_dtypes(include=[np.number])

corrformat = dfs.corr()
top_corr_features=corrformat.index
plt.figure(figsize=(15,15))
sns.heatmap(df[top_corr_features].corr(), annot=True, cmap="RdYlGn")
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
col_to_scale = ['FL_A','FL_B','FL_C','VS']
df[col_to_scale] = scaler.fit_transform(df[col_to_scale])
df.head()
```

#### Now splitting my dataset into train and test columns

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(df, test_size=0.2)
X_train = train[['FL_A', 'FL_C', 'VS']] 
y_train = train.Class

X_test = train[['FL_A', 'FL_C', 'VS']]
y_test = train.Class
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn_score = []
for k in range(1, 21):
    knn_classifier  = KNeighborsClassifier(n_neighbors=k)
    knn_classifier.fit(X_train, y_train)
    knn_score.append(knn_classifier.score(X_test, y_test))
```

```python
knn_score
```

```python
plt.figure(figsize=(30,30))
plt.plot([k for k in range(1,21)],knn_score,color="blue")
for i in range(1,21):
  plt.text(i, knn_score[i-1],(i,round(knn_score[i-1],4)))
plt.xticks([i for i in range(1,21)])
plt.xlabel("Number of Neighbors (K)",color="black",weight="bold",fontsize="18")
plt.ylabel("Scores",color="black",weight="bold",fontsize="18")
plt.title("K Neighbors Classifier scores for different K values",color="black",weight="bold",fontsize="20")
#plt.figure(figsize=(30, 20))
plt.show()
plt.rcParams["font.weight"]="bold"
plt.rcParams["axes.labelweight"]="bold"
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay

# create confusion matrix
KN_model = KNeighborsClassifier(8)
KN_model.fit(X_train,y_train)
y_pred = KN_model.predict(X_test)
cm = confusion_matrix(y_test, y_pred)

# plot confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
```

```python
from sklearn.metrics import classification_report, accuracy_score, recall_score, f1_score
print("Accuracy Score: ", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
TP = cm[1][1]
FP = cm[0][1]
TN = cm[0][0]
FN = cm[1][0]
b = ((TP+FP)*(TP+FN)*(TN+FP)*(TN+FN))
b = b**0.5
a = (TP*TN-FP*FN)
MCC_KNN = round(a/b,3 )
Precision_Score = TP / (FP + TP)
Recall_Score = TP / (FN + TP)
Accuracy_Score = (TP + TN)/ (TP + FN + TN + FP)
F1_Score = 2* Precision_Score * Recall_Score/ (Precision_Score + Recall_Score)
print("Accuracy  : ",Accuracy_Score*100)
print("Precision : ",Precision_Score*100)
print("Recall    : ",Recall_Score*100)
print("F1 score  : ",F1_Score*100)
print('MCC Score : ', MCC_KNN)
```
